# Poisson process
sample time equally

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%load_ext autoreload
%autoreload 1
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
# time interval must be far smaller than 1 
# because the jump interval is always 1.
def poisson_process(step, lam=1, time_interval=1e-3):
    r = np.random.rand(step)
    param = r < (np.exp(- lam * time_interval) * lam * time_interval)
    coord = np.cumsum(param)
    return coord
```

```python
step = 10000
lam = 1
time_interval = 1e-3

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)

t = np.arange(0, step * time_interval, time_interval)
x = poisson_process(step, lam, time_interval)
ax.plot(t, x)
plt.show()
```

```python
step = 100000
lam = 1
time_interval = 1e-3

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)

t = np.arange(0, step * time_interval, time_interval)
x = poisson_process(step, lam, time_interval)
ax.plot(t, x)
plt.show()
```

```python
step = 30000
time_interval = 1e-3
num = 20

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
for i in range(num):   
    t = np.arange(0, step * time_interval, time_interval)
    x = poisson_process(step)
    ax.plot(t, x)
plt.show()
```

### change parameters

```python
step = 100000
time_interval = 1e-3
lams = np.arange(0.2, 1.05, 0.2).round(2)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
for lam in lams:   
    t = np.arange(0, step * time_interval, time_interval)
    x = poisson_process(step, lam, time_interval)
    ax.plot(t, x, label=lam)
ax.legend()
plt.show()
```

# compensated Poisson process
$$ N(t) - \lambda t$$

```python
step = 10000
lam = 1
time_interval = 1e-3

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)

t = np.arange(0, step * time_interval, time_interval)
x = poisson_process(step, lam, time_interval)
x = x.astype(np.float64) - lam * t
ax.plot(t, x)
plt.show()
```
