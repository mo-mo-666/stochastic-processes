# Geometric Brownian motion
___
A *geometric Brownian motion* is a stochastic process satisfying the following stochastic differential equation:
$$ dS_t = \sigma S_t dB_t + \mu S_tdt,$$
and this solution is the following:
$$ S_t = S_0 \exp \big{(}\big{(}\mu - \frac{\sigma}{2}\big{)}t - \sigma B_t\big{)}.$$

This process is known as the **Brack-Scholes model** in finance.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

%matplotlib inline
%load_ext autoreload
%autoreload 1
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
def GBM(step, mu, sigma, time_interval=1e-3):
    nd = np.hstack((np.zeros(1), np.random.randn(step-1) * np.sqrt(time_interval)))
    gauss = np.cumsum(nd)
    geom = np.exp((mu - sigma**2 / 2) * time_interval * t + sigma * gauss)
    return geom
```

```python
step = 10000
mu = 0.05
sigma = 0.3
time_interval = 1e-3

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)

t = np.arange(0, time_interval * step, time_interval)
x = GBM(step, mu, sigma, time_interval)
ax.plot(t, x)
plt.show()
```

### change parameter

```python
step = 10000
mus = [0.02, 0.04, 0.045, 0.05, 0.055, 0.07]
sigma = 0.3
time_interval = 1e-3
num = 5

fig = plt.figure(figsize=(10, 5*num))
for i in range(num):
    ax = fig.add_subplot(num, 1, i+1)
    for mu in mus:
        t = np.arange(0, time_interval * step, time_interval)
        x = GBM(step, mu, sigma, time_interval)
        ax.plot(t, x, label=mu)
    ax.legend()
plt.show()
```
